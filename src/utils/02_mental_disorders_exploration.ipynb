```python
y = pd.DataFrame(df.iloc[:,-1])

y["Expert Diagnose"].unique() # array(['Bipolar Type-2', 'Depression', 'Bipolar Type-1', 'Normal']

y = pd.get_dummies(y,prefix=['Expert Diagnose'],
                   columns=['Expert Diagnose'],
                   drop_first=False)
```

```python
df['Sexual Activity'] = df['Sexual Activity'].str.split().str[0]
df['Concentration'] = df['Concentration'].str.split().str[0]
df['Optimisim'] = df['Optimisim'].str.split().str[0]
df.head()

for x in df['Sexual Activity']:
    x = int(x)

for x in df['Concentration']:
    x = int(x)

for x in df['Optimisim']:
    x = int(x)
```
